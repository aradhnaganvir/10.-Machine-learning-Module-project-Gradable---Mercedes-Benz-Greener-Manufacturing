# Apply label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
for col in New_columns:
 encoder.fit(df_train[col].append(df_test[col]).values)
 df_train[col] = encoder.transform(df_train[col])
 df_test[col] = encoder.transform(df_test[col])
```

# Perform dimensionality reduction. There are so many columns which can be reduced

```python
from sklearn.decomposition import PCA
```

```python
pcaModel = PCA(0.98, svd_solver='full')
X=df_train.drop('y', axis=1)
y=df_train['y']
pcaModel.fit(X)
```

```python
pcaModel.n_components_
```

# So in 12 columns we can accomate the 98% information, original datset contains 378 columns

```python
pcaModel.explained_variance_ratio_
```

```python
np.mean(pcaModel.explained_variance_ratio_)
```

# check if exp var ratio is greater than threshold value

```python
pcaModel.explained_variance_ratio_ > np.mean(pcaModel.explained_variance_ratio_) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
pca = PCA(n_components=9,svd_solver='full')
X_train = pca.fit_transform(X_train)
X_test = pca.fit_transform(X_test)
```

```python
pca_df_test = pca.fit_transform(df_test)
```
